```python
# ==========================================================
# CELLULE 1/2 - PREVIEW (AUCUN EXPORT)
# Fusion CPU LUT + FPGA, tableaux + graphes seulement
# ==========================================================
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

try:
    from IPython.display import display
except Exception:
    display = print

# ---------- CONFIG ----------
BASE_DIR = Path("Resultats validation SA")
CPU_DIR = BASE_DIR / "cpu_lut_only"
FPGA_DIR = BASE_DIR / "fpga_only"
OUT_DIR = BASE_DIR / "fusion_cpu_lut_vs_fpga"
OUT_DIR.mkdir(parents=True, exist_ok=True)

cpu_winner_path = CPU_DIR / "validation_cpu_lut_winner_results.csv"
fpga_winner_path = FPGA_DIR / "validation_fpga_winner_results.csv"
cpu_raw_path = CPU_DIR / "validation_cpu_lut_raw_results.csv"
fpga_raw_path = FPGA_DIR / "validation_fpga_raw_results.csv"

if not cpu_winner_path.exists():
    raise FileNotFoundError(f"Introuvable: {cpu_winner_path}")
if not fpga_winner_path.exists():
    raise FileNotFoundError(f"Introuvable: {fpga_winner_path}")

cpu_w = pd.read_csv(cpu_winner_path)
fpga_w = pd.read_csv(fpga_winner_path)

def add_snr_if_missing(df_w, raw_path):
    if "snr_db" in df_w.columns and df_w["snr_db"].notna().any():
        return df_w
    if not raw_path.exists():
        return df_w
    raw = pd.read_csv(raw_path)
    if not {"file", "method", "snr_db"}.issubset(raw.columns):
        return df_w
    snr_map = raw.groupby(["file", "method"], as_index=False)["snr_db"].first()
    out = df_w.merge(snr_map, on=["file", "method"], how="left", suffixes=("", "_raw"))
    if "snr_db_raw" in out.columns:
        out["snr_db"] = out.get("snr_db").fillna(out["snr_db_raw"])
        out = out.drop(columns=["snr_db_raw"], errors="ignore")
    return out

cpu_w = add_snr_if_missing(cpu_w, cpu_raw_path)
fpga_w = add_snr_if_missing(fpga_w, fpga_raw_path)

winner_results = pd.concat([cpu_w, fpga_w], ignore_index=True)
if "status" not in winner_results.columns:
    winner_results["status"] = "ok"

METHOD_LABELS = {"cpu_lut_angles": "CPU LUT angles", "fpga": "FPGA"}
ok = winner_results[winner_results["status"] == "ok"].copy()
if ok.empty:
    raise RuntimeError("Aucun résultat valide (status=ok).")

for col, default in [
    ("pd_flag", 0), ("pfa_flag", 0), ("pm_flag", 0),
    ("strict_success", 0), ("doppler_err_hz", np.nan), ("phase_err_chip", np.nan),
    ("peak_ratio", np.nan), ("winner_margin_db", np.nan), ("time_s", np.nan),
]:
    if col not in ok.columns:
        ok[col] = default

# ---------- Tableau global ----------
method_global = (
    ok.groupby("method", as_index=False)
    .agg(
        n_signals=("file", "count"),
        strict_success_pct=("strict_success", lambda s: 100.0 * np.nanmean(s)),
        Pd_pct=("pd_flag", lambda s: 100.0 * np.nanmean(s)),
        Pfa_pct=("pfa_flag", lambda s: 100.0 * np.nanmean(s)),
        doppler_mae_hz=("doppler_err_hz", lambda s: np.nanmean(np.abs(s))),
        phase_mae_chip=("phase_err_chip", lambda s: np.nanmean(np.abs(s))),
        peak_ratio_mean=("peak_ratio", "mean"),
        time_mean_ms=("time_s", lambda s: 1000.0 * np.nanmean(s)),
        time_median_ms=("time_s", lambda s: 1000.0 * np.nanmedian(s)),
        time_p95_ms=("time_s", lambda s: 1000.0 * np.nanpercentile(s.dropna(), 95) if s.dropna().size else np.nan),
    )
    .sort_values("method")
    .reset_index(drop=True)
)
method_global["method_label"] = method_global["method"].map(METHOD_LABELS).fillna(method_global["method"])

# ---------- Tableau par SNR ----------
if "snr_db" in ok.columns and ok["snr_db"].notna().any():
    snr_summary = (
        ok.groupby(["method", "snr_db"], as_index=False)
        .agg(
            n_signals=("file", "count"),
            strict_success_pct=("strict_success", lambda s: 100.0 * np.nanmean(s)),
            Pd_pct=("pd_flag", lambda s: 100.0 * np.nanmean(s)),
            Pfa_pct=("pfa_flag", lambda s: 100.0 * np.nanmean(s)),
            doppler_mae_hz=("doppler_err_hz", lambda s: np.nanmean(np.abs(s))),
            phase_mae_chip=("phase_err_chip", lambda s: np.nanmean(np.abs(s))),
            time_mean_ms=("time_s", lambda s: 1000.0 * np.nanmean(s)),
        )
        .sort_values(["snr_db", "method"])
        .reset_index(drop=True)
    )
else:
    snr_summary = pd.DataFrame(columns=["method","snr_db","n_signals","strict_success_pct","Pd_pct","Pfa_pct","doppler_mae_hz","phase_mae_chip","time_mean_ms"])
snr_summary["method_label"] = snr_summary["method"].map(METHOD_LABELS).fillna(snr_summary["method"])

# ---------- Concordance ----------
agreement_table = pd.DataFrame()
methods = sorted(ok["method"].dropna().unique().tolist())
if len(methods) >= 2:
    for c in ["strict_success", "prn_winner", "doppler_meas_hz", "phase_meas_chip"]:
        if c not in ok.columns:
            ok[c] = np.nan
    wide = ok.pivot(index="file", columns="method", values=["strict_success","prn_winner","doppler_meas_hz","phase_meas_chip"])
    rows = []
    for i, m1 in enumerate(methods):
        for m2 in methods[i+1:]:
            cols = [
                ("strict_success", m1), ("strict_success", m2),
                ("prn_winner", m1), ("prn_winner", m2),
                ("doppler_meas_hz", m1), ("doppler_meas_hz", m2),
                ("phase_meas_chip", m1), ("phase_meas_chip", m2),
            ]
            common = wide[cols].dropna()
            if common.empty:
                continue
            rows.append({
                "method_left": m1,
                "method_right": m2,
                "method_left_label": METHOD_LABELS.get(m1, m1),
                "method_right_label": METHOD_LABELS.get(m2, m2),
                "n_common": int(len(common)),
                "decision_agreement_pct": float(100.0 * np.mean(common[("strict_success", m1)] == common[("strict_success", m2)])),
                "prn_agreement_pct": float(100.0 * np.mean(common[("prn_winner", m1)] == common[("prn_winner", m2)])),
                "doppler_delta_mean_hz": float(np.mean(np.abs(common[("doppler_meas_hz", m1)] - common[("doppler_meas_hz", m2)]))),
                "phase_delta_mean_chip": float(np.mean(np.abs(common[("phase_meas_chip", m1)] - common[("phase_meas_chip", m2)]))),
            })
    agreement_table = pd.DataFrame(rows)

# ---------- Cas difficiles ----------
hard_cols = [
    "file","method","snr_db","prn_injected","prn_winner",
    "pd_flag","pfa_flag","pm_flag","doppler_true_hz","doppler_meas_hz","doppler_err_hz",
    "phase_true_chip","phase_meas_chip","phase_err_chip","peak_ratio","winner_margin_db","strict_success","time_s"
]
hard_cols = [c for c in hard_cols if c in ok.columns]
hard_cases = (
    ok.sort_values(["strict_success","winner_margin_db","peak_ratio","doppler_err_hz","phase_err_chip"],
                   ascending=[True,True,True,False,False], na_position="last")[hard_cols]
    .head(10).reset_index(drop=True)
)

# ---------- Affichage tableaux ----------
print("=== TABLEAU GLOBAL PAR METHODE ===")
display(method_global.round(4))

print("\n=== TABLEAU PAR SNR ===")
display(snr_summary.round(4) if not snr_summary.empty else snr_summary)

print("\n=== CONCORDANCE ENTRE METHODES ===")
display(agreement_table.round(4) if not agreement_table.empty else agreement_table)

print("\n=== CAS DIFFICILES ===")
display(hard_cases.round(4))

# ---------- Graphes ----------
if not snr_summary.empty:
    fig, axes = plt.subplots(1, 3, figsize=(15, 4.2))
    has_pfa = "Pfa_pct" in snr_summary.columns
    for m, g in snr_summary.groupby("method"):
        label = METHOD_LABELS.get(m, m)
        axes[0].plot(g["snr_db"], g["Pd_pct"], marker="o", label=label)
        if has_pfa:
            axes[0].plot(g["snr_db"], g["Pfa_pct"], marker="x", linestyle="--", label=f"{label} Pfa")
        axes[1].plot(g["snr_db"], g["doppler_mae_hz"], marker="o", label=label)
        axes[2].plot(g["snr_db"], g["phase_mae_chip"], marker="o", label=label)

    axes[0].set_title("Detection / Fausse alarme vs SNR")
    axes[0].set_xlabel("SNR (dB)")
    axes[0].set_ylabel("%")
    axes[0].grid(alpha=0.3)
    axes[0].legend(fontsize=8)

    axes[1].set_title("Doppler MAE vs SNR")
    axes[1].set_xlabel("SNR (dB)")
    axes[1].set_ylabel("Hz")
    axes[1].grid(alpha=0.3)
    axes[1].legend(fontsize=8)

    axes[2].set_title("Phase MAE vs SNR")
    axes[2].set_xlabel("SNR (dB)")
    axes[2].set_ylabel("chips")
    axes[2].grid(alpha=0.3)
    axes[2].legend(fontsize=8)

    plt.tight_layout()
    plt.show()

plt.figure(figsize=(7.2, 4.2))
plt.bar(method_global["method_label"], method_global["time_mean_ms"])
plt.title("Temps moyen par methode")
plt.ylabel("ms")
plt.grid(axis="y", alpha=0.3)
plt.tight_layout()
plt.show()

if not snr_summary.empty:
    runtime_by_snr = ok.groupby(["method","snr_db"], as_index=False)["time_s"].mean()
    runtime_by_snr["time_mean_ms"] = 1000.0 * runtime_by_snr["time_s"]
    has_pfa = "Pfa_pct" in snr_summary.columns

    for m in snr_summary["method"].dropna().unique():
        label = METHOD_LABELS.get(m, m)
        g = snr_summary[snr_summary["method"] == m].sort_values("snr_db")
        gt = runtime_by_snr[runtime_by_snr["method"] == m].sort_values("snr_db")

        fig, axes = plt.subplots(1, 4, figsize=(20, 4.4))
        axes[0].plot(g["snr_db"], g["Pd_pct"], marker="o", linewidth=2, label="Pd")
        if has_pfa:
            axes[0].plot(g["snr_db"], g["Pfa_pct"], marker="x", linestyle="--", linewidth=1.8, label="Pfa")
        axes[0].set_title(f"{label} - Detection")
        axes[0].set_xlabel("SNR (dB)")
        axes[0].set_ylabel("%")
        axes[0].grid(alpha=0.3)
        axes[0].legend(fontsize=8)

        axes[1].plot(g["snr_db"], g["doppler_mae_hz"], marker="o", linewidth=2)
        axes[1].set_title(f"{label} - Doppler MAE")
        axes[1].set_xlabel("SNR (dB)")
        axes[1].set_ylabel("Hz")
        axes[1].grid(alpha=0.3)

        axes[2].plot(g["snr_db"], g["phase_mae_chip"], marker="o", linewidth=2)
        axes[2].set_title(f"{label} - Phase MAE")
        axes[2].set_xlabel("SNR (dB)")
        axes[2].set_ylabel("chips")
        axes[2].grid(alpha=0.3)

        axes[3].plot(gt["snr_db"], gt["time_mean_ms"], marker="o", linewidth=2)
        axes[3].set_title(f"{label} - Temps moyen")
        axes[3].set_xlabel("SNR (dB)")
        axes[3].set_ylabel("ms")
        axes[3].grid(alpha=0.3)

        fig.suptitle(f"Validation fusionnee - {label}", fontsize=12)
        plt.tight_layout()
        plt.show()

print("\nPreview terminé (aucun export fichier).")
```

```python
# ==========================================================
# CELLULE 2/2 - EXPORT COMPLET (TABLEAUX + GRAPHES + DOCX)
# A lancer APRES la cellule 1/2
# ==========================================================
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# Verif variables issues de la cellule preview
required_vars = ["winner_results", "method_global", "snr_summary", "agreement_table", "hard_cases", "ok", "METHOD_LABELS", "OUT_DIR"]
missing = [v for v in required_vars if v not in globals()]
if missing:
    raise RuntimeError(f"Variables manquantes: {missing}. Exécute d'abord la cellule 1/2.")

OUT_DIR = Path(OUT_DIR)
OUT_DIR.mkdir(parents=True, exist_ok=True)

# ---------- Exports tableaux ----------
winner_results.to_csv(OUT_DIR / "fusion_winner_results.csv", index=False)
method_global.to_csv(OUT_DIR / "fusion_table_global.csv", index=False)
snr_summary.to_csv(OUT_DIR / "fusion_table_by_snr.csv", index=False)
agreement_table.to_csv(OUT_DIR / "fusion_table_method_agreement.csv", index=False)
hard_cases.to_csv(OUT_DIR / "fusion_table_hard_cases.csv", index=False)

# ---------- Exports graphes ----------
fig_paths = []

if not snr_summary.empty:
    fig, axes = plt.subplots(1, 3, figsize=(15, 4.2))
    has_pfa = "Pfa_pct" in snr_summary.columns
    for m, g in snr_summary.groupby("method"):
        label = METHOD_LABELS.get(m, m)
        axes[0].plot(g["snr_db"], g["Pd_pct"], marker="o", label=label)
        if has_pfa:
            axes[0].plot(g["snr_db"], g["Pfa_pct"], marker="x", linestyle="--", label=f"{label} Pfa")
        axes[1].plot(g["snr_db"], g["doppler_mae_hz"], marker="o", label=label)
        axes[2].plot(g["snr_db"], g["phase_mae_chip"], marker="o", label=label)

    axes[0].set_title("Detection / Fausse alarme vs SNR")
    axes[0].set_xlabel("SNR (dB)")
    axes[0].set_ylabel("%")
    axes[0].grid(alpha=0.3)
    axes[0].legend(fontsize=8)

    axes[1].set_title("Doppler MAE vs SNR")
    axes[1].set_xlabel("SNR (dB)")
    axes[1].set_ylabel("Hz")
    axes[1].grid(alpha=0.3)
    axes[1].legend(fontsize=8)

    axes[2].set_title("Phase MAE vs SNR")
    axes[2].set_xlabel("SNR (dB)")
    axes[2].set_ylabel("chips")
    axes[2].grid(alpha=0.3)
    axes[2].legend(fontsize=8)

    plt.tight_layout()
    p = OUT_DIR / "fig_global_snr_metrics.png"
    fig.savefig(p, dpi=220, bbox_inches="tight")
    fig_paths.append(p)
    plt.show()

fig = plt.figure(figsize=(7.2, 4.2))
plt.bar(method_global["method_label"], method_global["time_mean_ms"])
plt.title("Temps moyen par methode")
plt.ylabel("ms")
plt.grid(axis="y", alpha=0.3)
plt.tight_layout()
p = OUT_DIR / "fig_time_mean_by_method.png"
fig.savefig(p, dpi=220, bbox_inches="tight")
fig_paths.append(p)
plt.show()

if not snr_summary.empty:
    runtime_by_snr = ok.groupby(["method","snr_db"], as_index=False)["time_s"].mean()
    runtime_by_snr["time_mean_ms"] = 1000.0 * runtime_by_snr["time_s"]
    has_pfa = "Pfa_pct" in snr_summary.columns

    for m in snr_summary["method"].dropna().unique():
        label = METHOD_LABELS.get(m, m)
        g = snr_summary[snr_summary["method"] == m].sort_values("snr_db")
        gt = runtime_by_snr[runtime_by_snr["method"] == m].sort_values("snr_db")

        fig, axes = plt.subplots(1, 4, figsize=(20, 4.4))
        axes[0].plot(g["snr_db"], g["Pd_pct"], marker="o", linewidth=2, label="Pd")
        if has_pfa:
            axes[0].plot(g["snr_db"], g["Pfa_pct"], marker="x", linestyle="--", linewidth=1.8, label="Pfa")
        axes[0].set_title(f"{label} - Detection")
        axes[0].set_xlabel("SNR (dB)")
        axes[0].set_ylabel("%")
        axes[0].grid(alpha=0.3)
        axes[0].legend(fontsize=8)

        axes[1].plot(g["snr_db"], g["doppler_mae_hz"], marker="o", linewidth=2)
        axes[1].set_title(f"{label} - Doppler MAE")
        axes[1].set_xlabel("SNR (dB)")
        axes[1].set_ylabel("Hz")
        axes[1].grid(alpha=0.3)

        axes[2].plot(g["snr_db"], g["phase_mae_chip"], marker="o", linewidth=2)
        axes[2].set_title(f"{label} - Phase MAE")
        axes[2].set_xlabel("SNR (dB)")
        axes[2].set_ylabel("chips")
        axes[2].grid(alpha=0.3)

        axes[3].plot(gt["snr_db"], gt["time_mean_ms"], marker="o", linewidth=2)
        axes[3].set_title(f"{label} - Temps moyen")
        axes[3].set_xlabel("SNR (dB)")
        axes[3].set_ylabel("ms")
        axes[3].grid(alpha=0.3)

        fig.suptitle(f"Validation fusionnee - {label}", fontsize=12)
        plt.tight_layout()
        p = OUT_DIR / f"fig_method_{m}.png"
        fig.savefig(p, dpi=220, bbox_inches="tight")
        fig_paths.append(p)
        plt.show()

# ---------- Resume + paragraphe rapport ----------
def _get_row(df, method):
    g = df[df["method"] == method]
    return g.iloc[0] if not g.empty else None

def _fmt(x, nd=2):
    try:
        if pd.isna(x):
            return "N/A"
        return f"{float(x):.{nd}f}"
    except Exception:
        return str(x)

cpu = _get_row(method_global, "cpu_lut_angles")
fpga = _get_row(method_global, "fpga")

pair = agreement_table[
    ((agreement_table["method_left"] == "cpu_lut_angles") & (agreement_table["method_right"] == "fpga")) |
    ((agreement_table["method_left"] == "fpga") & (agreement_table["method_right"] == "cpu_lut_angles"))
]
pair = pair.iloc[0] if not pair.empty else None

if cpu is not None and fpga is not None and pd.notna(cpu["time_mean_ms"]) and pd.notna(fpga["time_mean_ms"]) and float(fpga["time_mean_ms"]) > 0:
    speedup_fpga_vs_cpu = float(cpu["time_mean_ms"]) / float(fpga["time_mean_ms"])
else:
    speedup_fpga_vs_cpu = np.nan

rapport_table = pd.DataFrame([{
    "n_signaux_cpu": int(cpu["n_signals"]) if cpu is not None else np.nan,
    "n_signaux_fpga": int(fpga["n_signals"]) if fpga is not None else np.nan,
    "succes_strict_cpu_pct": float(cpu["strict_success_pct"]) if cpu is not None else np.nan,
    "succes_strict_fpga_pct": float(fpga["strict_success_pct"]) if fpga is not None else np.nan,
    "pd_cpu_pct": float(cpu["Pd_pct"]) if cpu is not None else np.nan,
    "pd_fpga_pct": float(fpga["Pd_pct"]) if fpga is not None else np.nan,
    "pfa_cpu_pct": float(cpu["Pfa_pct"]) if cpu is not None else np.nan,
    "pfa_fpga_pct": float(fpga["Pfa_pct"]) if fpga is not None else np.nan,
    "doppler_mae_cpu_hz": float(cpu["doppler_mae_hz"]) if cpu is not None else np.nan,
    "doppler_mae_fpga_hz": float(fpga["doppler_mae_hz"]) if fpga is not None else np.nan,
    "phase_mae_cpu_chip": float(cpu["phase_mae_chip"]) if cpu is not None else np.nan,
    "phase_mae_fpga_chip": float(fpga["phase_mae_chip"]) if fpga is not None else np.nan,
    "temps_moy_cpu_ms": float(cpu["time_mean_ms"]) if cpu is not None else np.nan,
    "temps_moy_fpga_ms": float(fpga["time_mean_ms"]) if fpga is not None else np.nan,
    "speedup_fpga_vs_cpu": speedup_fpga_vs_cpu,
    "accord_decision_pct": float(pair["decision_agreement_pct"]) if pair is not None else np.nan,
    "accord_prn_pct": float(pair["prn_agreement_pct"]) if pair is not None else np.nan,
    "delta_doppler_moy_hz": float(pair["doppler_delta_mean_hz"]) if pair is not None else np.nan,
    "delta_phase_moy_chip": float(pair["phase_delta_mean_chip"]) if pair is not None else np.nan,
}])

rapport_table.to_csv(OUT_DIR / "fusion_table_rapport_resume.csv", index=False)

paragraphe = f"""
Dans cette campagne de validation, la comparaison entre la version CPU LUT angles et l'implémentation FPGA a été menée sur des jeux de signaux identiques afin d'évaluer simultanément la cohérence algorithmique et la performance d'exécution.

Les résultats montrent un taux de succès strict de {_fmt(cpu['strict_success_pct']) if cpu is not None else 'N/A'} % pour CPU LUT et de {_fmt(fpga['strict_success_pct']) if fpga is not None else 'N/A'} % pour FPGA, avec des probabilités de détection (Pd) respectives de {_fmt(cpu['Pd_pct']) if cpu is not None else 'N/A'} % et {_fmt(fpga['Pd_pct']) if fpga is not None else 'N/A'} %, tandis que les probabilités de fausse alarme (Pfa) restent de {_fmt(cpu['Pfa_pct']) if cpu is not None else 'N/A'} % (CPU) et {_fmt(fpga['Pfa_pct']) if fpga is not None else 'N/A'} % (FPGA). Les erreurs moyennes absolues sur l'estimation Doppler et phase sont respectivement de {_fmt(cpu['doppler_mae_hz']) if cpu is not None else 'N/A'} Hz et {_fmt(cpu['phase_mae_chip']) if cpu is not None else 'N/A'} chips pour CPU LUT, contre {_fmt(fpga['doppler_mae_hz']) if fpga is not None else 'N/A'} Hz et {_fmt(fpga['phase_mae_chip']) if fpga is not None else 'N/A'} chips pour FPGA.

L'analyse de concordance inter-méthodes indique un accord de décision de {_fmt(pair['decision_agreement_pct']) if pair is not None else 'N/A'} % et un accord PRN de {_fmt(pair['prn_agreement_pct']) if pair is not None else 'N/A'} %, avec des écarts moyens de {_fmt(pair['doppler_delta_mean_hz']) if pair is not None else 'N/A'} Hz sur le Doppler et {_fmt(pair['phase_delta_mean_chip']) if pair is not None else 'N/A'} chips sur la phase, ce qui confirme une forte cohérence fonctionnelle entre les deux approches.

Sur le plan temporel, le temps moyen est de {_fmt(cpu['time_mean_ms']) if cpu is not None else 'N/A'} ms pour CPU LUT et {_fmt(fpga['time_mean_ms']) if fpga is not None else 'N/A'} ms pour FPGA, soit une accélération FPGA d'environ {_fmt(speedup_fpga_vs_cpu)}x par rapport au CPU LUT. Ainsi, la version CPU LUT est pertinente comme référence logicielle de validation, tandis que l'implémentation FPGA constitue la solution privilégiée pour les campagnes volumineuses nécessitant une réduction du temps total de traitement.
""".strip()

with open(OUT_DIR / "fusion_conclusion_rapport.txt", "w", encoding="utf-8") as f:
    f.write(paragraphe)

# ---------- DOCX ----------
try:
    from docx import Document
    from docx.shared import Pt
except Exception:
    import sys, subprocess
    subprocess.check_call([sys.executable, "-m", "pip", "install", "python-docx"])
    from docx import Document
    from docx.shared import Pt

doc = Document()
style = doc.styles["Normal"]
style.font.name = "Times New Roman"
style.font.size = Pt(12)

doc.add_heading("Comparaison CPU LUT angles vs FPGA", level=1)
doc.add_paragraph("Synthèse automatique des résultats fusionnés (campagnes CPU LUT et FPGA).")
doc.add_heading("Tableau de synthèse", level=2)

rows = [
    ("Nombre de signaux CPU", _fmt(rapport_table.loc[0, "n_signaux_cpu"], 0)),
    ("Nombre de signaux FPGA", _fmt(rapport_table.loc[0, "n_signaux_fpga"], 0)),
    ("Succès strict CPU (%)", _fmt(rapport_table.loc[0, "succes_strict_cpu_pct"])),
    ("Succès strict FPGA (%)", _fmt(rapport_table.loc[0, "succes_strict_fpga_pct"])),
    ("Pd CPU (%)", _fmt(rapport_table.loc[0, "pd_cpu_pct"])),
    ("Pd FPGA (%)", _fmt(rapport_table.loc[0, "pd_fpga_pct"])),
    ("Pfa CPU (%)", _fmt(rapport_table.loc[0, "pfa_cpu_pct"])),
    ("Pfa FPGA (%)", _fmt(rapport_table.loc[0, "pfa_fpga_pct"])),
    ("Doppler MAE CPU (Hz)", _fmt(rapport_table.loc[0, "doppler_mae_cpu_hz"])),
    ("Doppler MAE FPGA (Hz)", _fmt(rapport_table.loc[0, "doppler_mae_fpga_hz"])),
    ("Phase MAE CPU (chip)", _fmt(rapport_table.loc[0, "phase_mae_cpu_chip"])),
    ("Phase MAE FPGA (chip)", _fmt(rapport_table.loc[0, "phase_mae_fpga_chip"])),
    ("Temps moyen CPU (ms)", _fmt(rapport_table.loc[0, "temps_moy_cpu_ms"])),
    ("Temps moyen FPGA (ms)", _fmt(rapport_table.loc[0, "temps_moy_fpga_ms"])),
    ("Accélération FPGA vs CPU (x)", _fmt(rapport_table.loc[0, "speedup_fpga_vs_cpu"])),
    ("Accord décision (%)", _fmt(rapport_table.loc[0, "accord_decision_pct"])),
    ("Accord PRN (%)", _fmt(rapport_table.loc[0, "accord_prn_pct"])),
    ("Delta Doppler moyen (Hz)", _fmt(rapport_table.loc[0, "delta_doppler_moy_hz"])),
    ("Delta phase moyen (chip)", _fmt(rapport_table.loc[0, "delta_phase_moy_chip"])),
]
tbl = doc.add_table(rows=1, cols=2)
tbl.style = "Table Grid"
tbl.rows[0].cells[0].text = "Métrique"
tbl.rows[0].cells[1].text = "Valeur"
for k, v in rows:
    r = tbl.add_row().cells
    r[0].text = str(k)
    r[1].text = str(v)

doc.add_heading("Paragraphe d'analyse", level=2)
doc.add_paragraph(paragraphe)

docx_path = OUT_DIR / "Rapport_Comparaison_CPU_LUT_vs_FPGA_Auto.docx"
doc.save(docx_path)

print("=== EXPORTS TERMINES ===")
print("- " + str(OUT_DIR / "fusion_winner_results.csv"))
print("- " + str(OUT_DIR / "fusion_table_global.csv"))
print("- " + str(OUT_DIR / "fusion_table_by_snr.csv"))
print("- " + str(OUT_DIR / "fusion_table_method_agreement.csv"))
print("- " + str(OUT_DIR / "fusion_table_hard_cases.csv"))
print("- " + str(OUT_DIR / "fusion_table_rapport_resume.csv"))
print("- " + str(OUT_DIR / "fusion_conclusion_rapport.txt"))
print("- " + str(docx_path))
print("\nGraphes exportés:")
for p in fig_paths:
    print("-", p)
```
